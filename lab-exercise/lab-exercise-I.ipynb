# 1. Python program to get the difference between a given number and 17, if the number is greater than 17, return double the abbsolute difference.

```python
A = float(input("Enter first number: "))
B = 17
C = A - B
if A > B:
    D = C * 2
    print("The answer is", D)
else: 
    D = abs(C)
    print("The answer is", D)
```

# 2. Write a python program  to calculate the sum of three given numbers, if the values are equall then return thrice os their sum.

```python
A = float(input("Enter first number: "))
B = float(input("Enter second number: "))
C = float(input("Enter third number: "))
if A == B == C:
    D = (A + B + C)*3
    print("The answer is", D)
else:
    D = A + B + C
    print("The answer is", D)
```

```python
A = int(input("Enter first number: "))
B = int(input("Enter second number: "))
C = int(input("Enter third number: "))

max(A, B, C)
T = max(A, B, C)
min(A, B, C)
U = min(A, B, C)
Q = (A + B + C) - max(A, B, C) - min(A, B, C)

print("The values in order are:", T, Q, U)
```

# 5. Write a Python function that takes a positive integer and returns the sum of the cube of all the positive integers smaller than the specified number.


```python
A = int(input("Enter number: "))
k = 0;
i = (range(1, A))
for h in i:
    z = h**3
    k = k + z
print(k)
```
